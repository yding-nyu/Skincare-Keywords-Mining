```python
def get_urls(url):
    """从每个列表页获取文章url"""
    req = Request(url= url, headers = headers)
    html = urlopen(req).read()
    soup = BeautifulSoup(html, features='lxml')
    all_href= soup.find_all('a', class_='m-cta--more')
    b_list = []
    for l in all_href:
        b_list.append('https://intothegloss.com'+l['href'])
    return b_list
```

```python
def get_info(url):
    """获取标题、作者、发布时间及正文内容"""
    time.sleep(0.1)
    req = Request(url, headers = headers)
    html = urlopen(req).read()
    soup = BeautifulSoup(html, features='lxml')
    
    if soup.find('h1', class_='c-h1 s-post-title m-visuals--borderTop') == None:
        return {'url': url, 'title': None, 'author and date': None, 'body': None}
    else:
        title = soup.find('h1', class_='c-h1 s-post-title m-visuals--borderTop').text
        a = soup.find_all('a',{'href': re.compile("/author/.*")})
        author_date = a[1].text
        body = soup.find('div', class_='s-post-text').text
        return {'url': url, 'title': title, 'author and date': author_date, 'body': body}
```

```python
def save_as_xls(info_list):
    """将获取到的文章信息存储到本地"""
    workbook = Workbook()
    worksheet = workbook.active
    worksheet.title = 'sheet1'
    titles = ['URL', 'Title', 'Author and Date', 'Content']
    keys = ['url', 'title', 'author and date', 'body']
    for index, title in enumerate(titles, start=1):
        worksheet.cell(row=1, column=index, value=title)
    for i, info in enumerate(info_list, start=1):
        for j,key in enumerate(keys, start=1):
            worksheet.cell(row=i+1, column=j, value=info[key])
    workbook.save('info.xlsx')
```

```python
list=[]
for i in range(1,30):
    list.append('https://intothegloss.com/sections/skincare/page/'+ str(i))
print(list)
```

```python
def main():
    """main function"""
    info_list = []
    for i in list:
        b_list = get_urls(i)
        for url in b_list:
            info_list.append(get_info(url))
    save_as_xls(info_list)
    
if __name__ == "__main__":
    main()
```
